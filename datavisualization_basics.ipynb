```python
import matplotlib.pyplot as plt
```

```python
# plotting lines using matplotlib
x_axis = [1,2,3]
y_axis = [4,5,6]

plt.plot(x_axis,y_axis, 'r')
```

```python
# title, labels and legend

x_axis = [1,2,3]
y_axis = [4,5,6]

plt.plot(x_axis,y_axis, 'g')

plt.title("My First Graph")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
```

```python
# 2 plots in one chart

x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]
x2_axis = [1,3,5,7]
y2_axis = [100,110,120,130]

plt.title("Two plots on one chart.")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")

plt.plot(x1_axis,y1_axis)
plt.plot(x2_axis,y2_axis)

plt.legend(["exponential", "linear"])
```

```python
# plotting Histograms 
# used to find out frequency distribution of the continues variable

values = [10,15,10,15,20]

plt.hist(values, color= 'y', width = 0.5)
```

```python
# plotting Bar Charts
# Bar plot is a diagramatic comparison of discret variables 

values = (10,15,20)
plt.bar(values, [0.5,1.0,1.5], width = 1.2)
```

```python
#plotting pie charts

values = [20,20,35,25]
plt.pie(values, labels=['Zitrone', 'pfel', 'Orangen', 'Birnen'], explode=[0,0,0.1,0])
plt.title("Pie Chart")
```
